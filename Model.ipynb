```python
# Import libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import data.

us_states_df = pd.read_csv('C:\\Users\\jtmar\\dev\\Data Science Work\\covid-19-data\\us-states.csv')
us_counties_df = pd.read_csv('C:\\Users\\jtmar\\dev\\Data Science Work\\covid-19-data\\us-counties.csv')
us_df = pd.read_csv('C:\\Users\\jtmar\\dev\\Data Science Work\\covid-19-data\\us.csv')

display(us_states_df.head())
display(us_counties_df.head())
display(us_df.head())
```

```python
us_states_deaths_df = us_states_df.pivot(index = 'date', columns = 'state', values = 'deaths').fillna(0)
us_states_cases_df = us_states_df.pivot(index = 'date', columns = 'state', values = 'cases').fillna(0)
```

```python
(100*us_states_deaths_df/us_states_cases_df).plot(rot = 45, title = 'Percent Case Mortality', figsize = (17,17))
```

```python
(us_states_cases_df/100000).plot(y='Utah',rot = 45, title = 'Cases per 100000', figsize = (18,5))
```

```python
state = 'Utah'
(us_states_deaths_df[state] - us_states_deaths_df[state].shift(periods = 1).fillna(0)).plot(rot = 45, title = 'New Deaths Per Day', figsize = (18,5), kind='line')
```

```python
state = 'Utah'
(us_states_cases_df[state] - us_states_cases_df[state].shift(periods = 1).fillna(0)).plot(rot = 45, title = 'New Cases Per Day', figsize = (18,5), kind='line')
```
